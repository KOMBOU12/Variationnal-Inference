```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Jan 16 10:21:30 2024

@author: solange
"""

import skgstat as skg
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
skg.plotting.backend('matplotlib')

### Charger les donnes

data1 = skg.data.pancake(N=180, seed=42, as_dataframe=True).get('sample')

### Reprsenter les donnes

fig, ax = plt.subplots()
cax = ax.scatter(data1.x,data1.y,c=data1.v)
ax.set_title('Gaussian noise with vertical colorbar')

cbar = fig.colorbar(cax)
labels = cbar.ax.get_yticklabels()
labels[0].set_verticalalignment('top')
labels[-1].set_verticalalignment('bottom')

### Calculer et reprsenter le variogramme

V1 = skg.Variogram(data1[['x', 'y']].values, data1.v.values, maxlag='median', normalize=False)
V1.bin_func='even' # 'uniform
V1.model = 'spherical' # 'exponential' 'gaussian'
V1.plot(show=False)
print(V1)

```
